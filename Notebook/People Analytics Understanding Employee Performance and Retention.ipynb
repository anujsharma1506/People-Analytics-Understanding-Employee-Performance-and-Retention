### Handling Missing Data

```python
# Handling missing values

# Fill missing 'ExitDate' with 'Not Applicable' for active employees
employee_data['ExitDate'].fillna('Not Applicable', inplace=True)

# Fill missing 'TerminationDescription' with 'Not Applicable' for active employees
employee_data['TerminationDescription'].fillna('Not Applicable', inplace=True)

# Confirm there are no missing values left
missing_values_after = employee_data.isnull().sum()
```

```python
# Handling missing values in 'DOB'
# For simplicity, we'll fill missing 'DOB' with the median DOB of non-missing values
median_dob = employee_data['DOB'].median()
employee_data['DOB'].fillna(median_dob, inplace=True)

# Recalculate 'Age' after filling missing 'DOB'
employee_data['Age'] = pd.Timestamp.now().year - employee_data['DOB'].dt.year
```

```python
print("Missing Values After Handling:\n", missing_values_after)
```

```python
employee_data['Retention'] = employee_data['ExitDate'] == 'Not Applicable'

# Grouping by factors of interest and calculating retention rates
factors = ['BusinessUnit', 'GenderCode','RaceDesc',  'Performance Score']
retention_analysis = {}

for factor in factors:
    retention_analysis[factor] = (
        employee_data.groupby(factor)['Retention']
        .mean()
        .sort_values(ascending=False)
        .reset_index()
    )
```

### Retention Rates by Business Unit

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=retention_analysis['BusinessUnit'], x='Retention', y='BusinessUnit', palette='coolwarm')
plt.title("Retention Rates by BusinessUnit")
plt.xlabel("Retention Rate")
plt.ylabel("BusinessUnit")
plt.show()
```

### Findings:
1. Business units such as `PL` and `CCDR` have higher retention rates (~52%).
2. Units like `SVG` and `WBL` exhibit lower retention rates (~45%).

### Insights:
1. Business units with specialized operations or fewer external influences (e.g., `PL`) tend to retain employees better.
2. Units with lower retention could benefit from targeted engagement programs to address employee concerns.


### **Retention Rates by Performance Score**

```python
plt.figure(figsize=(7, 6))
sns.barplot(data=retention_analysis['Performance Score'], x='Performance Score', y='Retention', palette='coolwarm')
plt.title("Retention Rates by Performance Score")
plt.xlabel("Retention Rate")
plt.ylabel("Performance Score")
plt.show()
```


### Findings:
1. Employees with "Fully Meets" and "Needs Improvement" scores have slightly higher retention (~49%).
2. Those on "PIP" (Performance Improvement Plan) have lower retention (~47%).

### Insights:
1. Employees meeting expectations are more likely to stay, highlighting the importance of realistic performance benchmarks.
2. Poor retention of "PIP" employees suggests a need for better support mechanisms during performance improvement phases.

```python
# Analysis of "Current Employee Rating" across Job Titles, Departments, and Business Units

# Factors of interest
rating_factors = ['Title', 'DepartmentType', 'BusinessUnit']
rating_analysis = {}

for factor in rating_factors:
    # Grouping by the factor and calculating the mean rating
    rating_analysis[factor] = (
        employee_data.groupby(factor)['Current Employee Rating']
        .mean()
        .sort_values(ascending=False)
        .reset_index()
    )



```

## Average Employee Rating by Job Title

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=rating_analysis['Title'],
    x='Current Employee Rating',
    y='Title',
    palette='viridis'
)
plt.title("Average Current Employee Rating by Job Title")
plt.xlabel("Average Employee Rating")
plt.ylabel("Job Title")
plt.show()
```

### Findings:
1. The highest ratings are for `IT Manager - Infra` (3.55), `Director of Operations` (3.45), and `Software Engineering Manager` (3.40).
2. The lowest ratings are for `President & CEO` (1.91), `IT Manager - Support` (2.36), and `Director of Sales` (2.36).

### Insights:
1. Leadership roles like `IT Manager - Infra` and `Director of Operations` have higher satisfaction, indicating strong alignment with their job responsibilities.
2. Lower ratings for executive-level positions like `President & CEO` suggest potential dissatisfaction or higher stress levels in top management roles.

## Average Employee Rating by Department Type

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    data=rating_analysis['DepartmentType'],
    x='Current Employee Rating',
    y='DepartmentType',
    palette='viridis'
)
plt.title("Average Current Employee Rating by Department Type")
plt.xlabel("Average Employee Rating")
plt.ylabel("Department Type")
plt.show()
```

### Findings:
1. The `Admin Offices` department has the highest average rating (3.03), followed by `Production` (2.98).
2. The lowest ratings are in `Executive Office` (2.79) and `Software Engineering` (2.90).

### Insights:
1. Employees in administrative and production departments may experience more structured and predictable workflows, leading to higher satisfaction.
2. Lower ratings in executive and software engineering departments could point to challenges like demanding workloads or lack of clear progression opportunities.


### Analysis of termination types and reasons across business units 

```python


# Grouping by TerminationType and TerminationDescription
termination_counts = (
    employee_data.groupby(['TerminationType', 'TerminationDescription'])
    .size()
    .sort_values(ascending=False)
    .reset_index(name='Count')
)

# Grouping by BusinessUnit and TerminationType
termination_business_unit = (
    employee_data.groupby(['BusinessUnit', 'TerminationType'])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

# Grouping by JobFunctionDescription and TerminationType
termination_job_function = (
    employee_data.groupby(['JobFunctionDescription', 'TerminationType'])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

# Display the top termination types and reasons
print("\nMost Common Termination Types and Reasons:\n", termination_counts.head(10))



```

### Termination Types Across Business Units

```python
# Visualization of termination types across business units
termination_business_unit.set_index('BusinessUnit').plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')
plt.title("Termination Types Across Business Units")
plt.xlabel("Business Unit")
plt.ylabel("Count")
plt.legend(title="Termination Type", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

## Findings:
1. Unknown Terminations:
   - The "Unk" termination type dominates all business units, highlighting inconsistent or incomplete documentation of termination reasons.
2. Voluntary Terminations:
   - Voluntary terminations (resignations and retirements) are relatively balanced across business units but slightly higher in `SVG` and `WBL`.

## Insights:
1. Improve Documentation:
   - The high prevalence of "Unknown" termination types indicates a need for better documentation processes to capture accurate reasons for employee departures.
2. Retention Strategies:
   - Units like `SVG` and `WBL` with higher voluntary terminations could benefit from enhanced retention initiatives, such as career development programs and employee engagement activities.
